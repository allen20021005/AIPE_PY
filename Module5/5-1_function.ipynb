```python
# 建立 function 1 # 沒有 return 值 沒有參數
def greeting():
    print("Hello,World")

greeting()
greeting()
greeting()
```

```python
# 建立 function 2  # 有return 值的 function
def hello_texts():
    return "Good Mornings"
hello = hello_texts()
print(hello)
```

```python
# 建立 function 3
def hello_user(name):
    print(f"hello,{name}")

hello_user("CBKKKK")
```

```python
# 建立 function 4
def say_hello(name,greeting="Good Morning!!"):  # 要預設值的參數一定要放在後面
    print(f"{greeting} {name}")

say_hello("Brain")
```

```python
# 改寫 Module2 practice
totals = [5600,3400,9000,4000]
def get_discount(total:int):
    if total >= 2000:
        total -= 200*(total//2000)
    return total
for total in totals:
    result = get_discount(total)
    print(result)
```
