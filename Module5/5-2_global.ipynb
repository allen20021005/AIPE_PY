```python
# 全域變數
name = "Allen"

# 定義函式
def say_hello():
    name = "Alex"   #區域變數 不會影響到全域變數

    print(f"Hello {name}")

# 執行函式
say_hello()

# 輸出 name
print(name)
```

```python
# 全域變數
name = "Allen"

# 定義函式
def say_hello():
    # function內部沒有變數，直接使用全域變數
    print(f"Hello {name}")

# 執行函式
say_hello()

# 輸出 name
print(name)
```

##### 使用 global 讓函式存取全域變數的值

```python
# 全域變數
name = "Allen"

# 定義函式
def say_hello():
    global name     #加上 global 對全域變數重新賦值
    name = "Alex"   

    print(f"Hello {name}")

# 執行函式
say_hello()

# 輸出 name
print(name)
```
