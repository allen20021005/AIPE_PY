```python
fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['green', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': ['orange'] # 錯誤的寫法
    }
}
print(fruit_dict)
```

```python
# 使用 for 迴圈讀取 dict 資料1 -->只回傳Key
for key in fruit_dict:
    print(key)
```

```python
# 使用 for 迴圈讀取 dict 資料2 -->dict.values() 只回傳 value
for value in fruit_dict.values():
    print(value)
```

```python
# 使用 for 迴圈讀取 dict 資料3 --> key , value 都回傳 dict.items()
for key, value in fruit_dict.items():
    print(key,value)
```

```python
for key, value in fruit_dict.items():
    for color in value["color"]:
        print(f"{key}有{color}")
```

```python
for key, value in fruit_dict.items():
    result = ','.join(value['color'])
    print(f"{key} 有 {result}")
```

```python
# 取得 dict 的 value
print(fruit_dict.get('kiwi'))
if fruit_dict.get('kiwi') == None:
    print(f'kiwi 不在 fruit_dict 裡面')
```

```python
# 資料處理
fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'}]
```

```python
# 將資料整理成 {'apple': ['red', 'green'], 'grape': ['purple', 'green'], 'kiwi': ['yellow', 'green']}
result ={}
for fruit_color in fruit_color_list:
    for fruit,color in fruit_color.items():
        if result.get(fruit) == None:  #還沒有建立過該key值
            result[fruit]=[color]
        else:
            result[fruit].append(color)

print(result)    
```

##### `defaultdict(<type>)`

```python
# 當訪問的鍵不存在時，可以自動創建鍵和預設值，而不是拋出KeyError。
from collections import defaultdict

# 以int型態的預設值 0 來初始化
result = defaultdict(int)

print(result['apple'])
print(result)
```

```python
from collections import defaultdict


fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'}]
result = defaultdict(list)

for fruit_color in fruit_color_list:
    for fruit,color in fruit_color.items():
        result[fruit].append(color)

result = dict(result) #強制轉換成一般的dict
print(result)
```
