```python
# f-string (又稱作 formatted string literals, version >= 3.6)
msg = "Hello World"

print(f"[msg] {msg} !")

name = "CBK"
age = 25

print(f'{name} is {age} years old')
```

```python
# 靠左對齊、靠右對齊
data = "Hello"
# 要靠右加 > 符號

print(f'[{data:10}]')
print(f'[{data:>10}]')

```

```python
# 靠左對齊、靠右對齊，字數不足，補「ㄏ」
# (沒寫補什麼，預設空格)
data = "Hello"
print(f'[{data:_<10}]')
print(f'[{data:X>10}]')
```

```python
# 置中對齊
data = "Hello"
print(f'[{data:^10}]')
```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)
data = "Hello"
print(f'[{data:.2}]')

```

```python
user_id = "A123456789"
print(f"{user_id:.4}******")
print(f"{user_id:*<10.4}")
num = 3.14256
print(f"{num:.2f}")
total = 168321684
print(f"{total:,}")
```
