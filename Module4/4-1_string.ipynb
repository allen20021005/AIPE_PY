```python
# 字串變數初始化
data = "Hello,World"
print(data)
print(data[0]) #Python由0起算
```

```python
data = "Hell" + "o"*5 + "," + "World!!"
print("="*20)
print(data)
print("="*20)
```

```python
# 替換字串
'''
str,replace(<old>,<new>)
將 string 中的 str1 替換成 str2
'''
name = "Alex"
#,前面是舊的，後面是要更改的
new_name = name.replace("ex","len") # 會retunr 一個新的資料，而不是改變原本的變數 -> 所以要用一個變數去接回傳的值
print(new_name)
```

```python
# 去除兩側空格
'''
str.strip()
去除字串 string 左、右兩邊的空格
'''
msg = "___Hello,World__"

result = msg.strip("_")
print(result)
```

```python
# 字串小寫變大寫
'''
str.lower()

將字串 string 裡的字母全部改成小寫
'''
msg = "HELLO"
result = msg.lower()
print(result)

'''
str.upper()
將字串 string 裡的字母全部改成大寫
'''
name = "cbk"
result = name.upper()
print(result)
```

```python
# 字串分割成 list -> str.split()
greeting = "hello world! hello John"

word_list = greeting.split("!")
print(word_list)
```
