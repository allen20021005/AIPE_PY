```python
# 初始化 dict
user_info ={"name":"CBK","age":30}
print(user_info)
```

```python
# 印出 蘋果的價格
fruit_price ={
    "apple":40,
    "orange":50,
    "tomato":100
}
print(fruit_price)
print(fruit_price["apple"])
```

```python
# 修改 橘子的價格
fruit_price["orange"] = 150
print(fruit_price)
```

```python
# 刪除 番茄
fruit_price ={
    "apple":40,
    "orange":50,
    "tomato":100
}
fruit_price.pop("tomato")
print(fruit_price)
fruit_price ={
    "apple":40,
    "orange":50,
    "tomato":100
}
del fruit_price["tomato"]
print(fruit_price)
```

```python
# 建立一個 value 含有 dict 的 dict
user_info = {
    "name" : "CBK",
    "age" : 30,
    "info" :{
        "nickname": "BBBBBK",
        "role":"Pochita",
        "phones":["0912345678","0987654321"]
    }
}
print(user_info)
```

```python
# 取得 nickname
print(user_info["info"])
print(user_info["info"]["nickname"])
```

```python
# 取得所有 phone_number
print(user_info["info"]["phones"][1])
```

```python
'''
改寫 fruit_dict 每個水果需要 price, quantity/amount, color
apple 有 綠色 和 紅色
orange 有 橘色
tomato 有 紅色
'''
fruit_dict = {
    "apple":{
        "price" : 50,
        "quantity" : 30,
        "amount" : 20,
        "color" : ["green","red"]
    },
    "onange":{
        "price" : 40,
        "quantity" : 70,
        "amount" : 10,
        "color" : ["orange"]
    },
    "tomato":{
        "price" : 10,
        "quantity" : 40,
        "amount" : 90,
        "color" : ["red"]
    }
}
print(fruit_dict)
```

##### dict 的合併

```python
# dict 的合併
#資料合併前提是要資料結構是一樣的

data = {
    "banana":{
        "price":20,
        "quantity":50,
        "color":["yellow"]
    }
}

fruit_dict.update(data)
print(fruit_dict)
```

##### 另外的 print 方式 `pprint` (pretty-print) 

```python
# 從 Module 引入 function
from pprint import pprint
pprint(fruit_dict) #注意順序會被打亂
print("="*80)
pprint(fruit_dict,sort_dicts=False)
```

```python
'''
新增顏色
apple 有新的品種 顏色要有黃色
tomato 有綠色
'''
print(fruit_dict["apple"]["color"])
fruit_dict["apple"]["color"].append("yellow")
fruit_dict["tomato"]["color"].append("green")

pprint(fruit_dict,sort_dicts=False)
```
