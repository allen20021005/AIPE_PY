```python
# 多組文字
msg = '%s, %s!' % ('Hello','World')
print(msg)
```

```python
# 整數
price = 50
#放字串會導致錯誤
msg = "蘋果 %d 元" % (price)
print(msg)
```

```python
# 文字與整數
name = "CBK"
age = 25

result = "%s is %d years old." % (name,age)
print(result)
```

```python
# 指定寬度 (維持 10 個字元長度，預設向右看齊)
msg = "[%10s]" % ("Hello")
print(msg)
#預設靠右對齊
```

```python
# 靠左對齊 (維持 10 個字元長度，向左對齊)
msg = "%-10s" % ("Hello")
print("="*13)
print("||" + msg + "||")
print("="*13)

```

```python
# 指定浮點數位數
result = "%.2f" % (3.1415)
print(result)
```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)
result = "%.3s" % ("Hello")
print(result)
```

```python
# 空白補 0
result = "[%08.2f]" % (3.314566)
print(result)
```
