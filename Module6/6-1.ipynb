```python
# 類別名稱
class Car:
# 類別屬性 (Class Attribute)
    wheels = 4 
    def __init__(self,brand,model,price):
        #實例屬性
        self.brand = brand
        self.model = model
        self.price = price
        print("已建立一個物件")
    # 方法 (Method)
    def display_info(self):
        print(f'{self.brand}-{self.model} has {self.wheels} wheels.')
    def show_discount_price(self,discount):
        print(f'{self.brand}-{self.model} price is ${int(discount*self.price):,}')
print(Car.wheels)

car1 = Car("toyota","camery",800000)   #建立 Car class 的一個實例
car2 = Car("Benz","C300",1000000)

print(car1.wheels)
print(car1.brand)
print(car1.model)

print(car2.brand)
print(car2.wheels)

car1.display_info()
car2.display_info()

car1.show_discount_price(0.8)
```
