##### try...except...

```python
# 可以連續定義數個例外處理機制
# 最符合的例外事件類別會先拋出
try:
    x = 10
    print(y)
    print(X)    #上面有抱錯 不執行該行程式
except: # 捕捉所有的錯誤
    print("發生錯誤")
```

```python
try:
    x = 10
    print(x[0])
    print(y)
    print(X)    
except NameError: # 捕捉特定錯誤類別
    print("發生NameError")
except TypeError:
    print("發生TypeError")
```

```python
try:
    x = 10
    print(x[0])
    print(y)
    print(X)    
    # as -> 把 Exception 產生的資料賦值給 e     e = 錯誤訊息
except Exception as e : #捕捉大部分錯誤類別 # 一律放在最下方
    print("發生錯誤")   
    print(e)
```

```python
# 多層例外處理:
try:
    x = 10
    try:  # 不影響原本程式 單獨使用
        print(y)
    except NameError as e:
        print(e)
except Exception as e:
    print(e)
```
