```python
# 匯入套件
import os

#列出當前資料夾的檔案
files = os.listdir()
print(files)

# 檔案路徑
file_path = 'file.txt'

print(os.path.exists(file_path))

# 如果指定路徑的檔案存在，則執行刪除
if os.path.exists(file_path):
    os.remove(file_path)
    print(f'{file_path} 已刪除')

```

```python
# 建立資料夾
import json
folder ="Test"
if not os.path.exists(folder):
    os.mkdir(folder)
with open("data.json","w") as f:
    json.dump([1,2,3],f,indent =4)
```
