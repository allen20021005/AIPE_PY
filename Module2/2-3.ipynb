整數 (int)、浮點數 (float)

```python
num = 20
float_num = 12.5
print(num)
print(float_num)

# 複製該行程式碼 → Alt + Shift + 方向鍵
# 移動該行程式碼 → Alt + 方向鍵
# 打開/關閉終端機 → Ctrl + `
# 移動到終端機 → Ctrl + `
# 移動到某個 window → Ctrl + 數字鍵
# 關閉 window → Ctrl + W
# 同時改多個變數 → Ctrl + D
```

布林值 (boolean)

```python
parse_ok = True
parse_not = False

print(parse_ok)
print(parse_not)
```

字串 (string)

```python
name = "CBK"
print(name)

data = "False"
print(data)
#檢查資料型態 type()
print(type(data))
```

列表 (list)

```python
num_list = [1,2,3]
print(num_list)
print(type(num_list))
```

元組 (tuple)

```python
num_tuple = (1,2,3)
print(num_tuple)
print(type(num_tuple))
```

集合 (set)

```python
 #不會有重複的值，有的話只會顯示一個
num_set = {1,2,3,3}
print(num_set)
```
