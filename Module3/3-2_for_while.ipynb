| 用法 | 說明 |
|:---:|:---:|
| `range(m)` | 走訪 0 到 m-1 的數字 |
| `range(n, m)` | 走訪 n 到 m-1 的數字  |
| `range(n, m, step)` | 以step為間距，走訪 n 到 m-1 的數字 |

##### for

```python
# range(n)
for i in range(5):
    print(i,end=",")
```

```python
# range(n, m)
for i in range(1,11):
    print(i,end=",")
```

```python
# range(n, m-1, step)
for i in range(1,16,2):
    print(i,end=",")
```

### 小試身手 for  
請產出 a 開頭的 n 個偶數(含 a 本身)  
Ex:  
a = 5, n = 3  
Output : 6 8 10

a = 4, n = 5  

Output : 4 6 8 10 12

```python
a = int(input())
n = int(input())

if a % 2 != 0: 
    a+=1
for i in range(n):
    print(a + i*2)
```

```python
# 嘗試產出
"""
* * * * *
 * * * *
  * * *
   * *
    *
"""
num = 5
for i in range(5,0,-1):    
    print(""*(num-i) + "*"*i)


# 嘗試產出
"""
    *
   * *
  * * *
 * * * *
* * * * *
"""
```
