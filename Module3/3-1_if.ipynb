條件語句 if-elif-else

```python
'''
條件判斷
if ...
if ... else ...
if ... elif ... else ...

'''
```

##### if

```python
age = 20

if age == 20:
    print("條件成立")
```

##### if...else...

```python
score = 59

if score >= 60:
    print("及格")
else:
    print("不及格")
```

##### if...elif... (else)

```python
score = 100
if score == 100:
    print("滿分")
elif score >= 90:
    print
```
